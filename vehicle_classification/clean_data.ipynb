```python
from scipy.io import loadmat
import pathlib
import shutil
import os
```

### 1. Create Folders

```python
label = loadmat('raw/devkit/cars_meta.mat')
class_names = label['class_names'][0]
for class_name in class_names:
    print('--', class_name[0], '--')
    class_name[0] = class_name[0].replace("/","_")
    path_train = 'data/train/'+class_name[0]
    path_test = 'data/test/'+class_name[0]
    path_val = 'data/valid/'+class_name[0]
    pathlib.Path(path_train).mkdir(parents=True, exist_ok=True)
    pathlib.Path(path_test).mkdir(parents=True, exist_ok=True)
    pathlib.Path(path_val).mkdir(parents=True, exist_ok=True)
```

### 2. Copy Images to Folders

```python
def copy_images(mat_path, raw_path, new_path):
    train_label = loadmat(mat_path)
    classes = train_label['annotations'][0]
    for i in range(len(classes)):
        class_name_index = classes[i]['class'][0][0] - 1
        folder_name = class_names[class_name_index][0]
        file_name = classes[i]['fname'][0]
        shutil.copyfile(raw_path+file_name, new_path+folder_name+'/'+file_name)
        print((i+1), class_names[class_name_index][0])

copy_images('raw/devkit/cars_train_annos.mat', 'raw/cars_train/', 'data/train/')
copy_images('raw/cars_test_annos_withlabels.mat', 'raw/cars_test/', 'data/valid/')
```

#### 2.1 Move Test to Train (add more data to training set)

```python
import math
import os
path_train = 'data/train/'
path_valid = 'data/valid/'
folders = os.listdir(path_valid)
for folder_name in folders:
    if os.path.isdir(path_valid+folder_name):
        full_folder_name = path_valid+folder_name
        
        num_images = len([name for name in os.listdir(full_folder_name) if os.path.isfile(os.path.join(full_folder_name, name))])
        num_val = math.floor(num_images/2.5)
        print(num_val)
        for i, file in enumerate(os.listdir(full_folder_name)):
            if i < num_val:
                file_valid = path_valid+folder_name+'/'+file
                file_train = path_train+folder_name+'/'+file
                print(file_valid, file_train)
                shutil.move(file_valid, file_train)
                
```

### 3. Merge to Each Type (eg. sedan, convertable, van...)

```python
file = open('class_data.rtf', 'r')
text = file.read().strip()
file.close()
text = text.replace('{\\rtf1\\ansi\\ansicpg1252\\cocoartf1504\\cocoasubrtf830\n{\\fonttbl\\f0\\fswiss\\fcharset0 Helvetica;}\n{\\colortbl;\\red255\\green255\\blue255;}\n{\\*\\expandedcolortbl;;}\n\\margl1440\\margr1440\\vieww10800\\viewh8400\\viewkind0\n\\pard\\tx720\\tx1440\\tx2160\\tx2880\\tx3600\\tx4320\\tx5040\\tx5760\\tx6480\\tx7200\\tx7920\\tx8640\\pardirnatural\\partightenfactor0\n\n\\f0\\fs24 \\cf0 ','')
text = text[:-1]
text = text.replace('\\\n', '\n')
text_array = text.split('\n\n')

def move_to_type_by_data_set(data_folder):
    temp_folder = 'data/temp/'
    for group_text in text_array:
        vehicle_types = group_text.split('\n')
        head = vehicle_types.pop(0).replace('###', '')

        if head != 'other': # other type is very confusing so exclude it
            pathlib.Path(temp_folder+head).mkdir(parents=True, exist_ok=True)
            for folder in vehicle_types:
                for file in os.listdir(data_folder+folder):
                    shutil.move(data_folder+folder+'/'+file, temp_folder+head+'/'+folder+file)
    shutil.rmtree(data_folder)
    shutil.move(temp_folder, data_folder)

move_to_type_by_data_set('data/train/')
move_to_type_by_data_set('data/valid/')
```
